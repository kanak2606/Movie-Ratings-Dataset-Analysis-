# Problem 1 (Imputing the mean)

```python
# Impute missing ratings in place of NAN by using the mean of row and column
col_mean = np.nanmean(df, axis=0)
row_mean = np.nanmean(df, axis=1)
df = df.fillna(pd.Series(row_mean, index=df.index))
df = df.fillna(pd.Series(col_mean, index=df.columns))
```

```python
# Total number of Null values after imputing
df.isnull().sum()
```

# Problem 2 (Generating 399 Models)

```python
# Compute COD for each movie
cods = []
for i in range(400):
    # Select the ratings for movie i and the ratings for other movies
    ratings = df.iloc[:, i]
    other_ratings = df.drop(df.columns[i], axis=1)

    # Train a linear regression model on the other ratings
    lr = LinearRegression()
    lr.fit(other_ratings, ratings)

    # Predict the ratings for the movie i
    y_pred = lr.predict(other_ratings)

    # Compute the COD for the linear regression model
    cod = r2_score(ratings, y_pred)
    cods.append(cod)
```

```python
# Report the average COD and generate a histogram
avg_cod = np.mean(cods)
print("Average COD:", avg_cod)
import matplotlib.pyplot as plt
plt.hist(cods, bins=20)
plt.title("Distribution of COD")
plt.xlabel("COD")
plt.ylabel("Frequency")
plt.show()
```

```python
# Identify the movies that are most easily and hardest to predict
cods_dict = {i: cod for i, cod in enumerate(cods)}
sorted_cods = sorted(cods_dict.items(), key=lambda x: x[1], reverse=True)
print("Top 10 movies that are most easily predicted:")
print("Movie\tCOD\tBest predictor")
for i in range(10):
    movie = sorted_cods[i][0]
    cod = sorted_cods[i][1]
    best_predictor = df.corrwith(df.iloc[:, movie]).sort_values(ascending=False).index[1]
    print(f"{movie}\t{cod:.4f}\t{best_predictor}")
print("\nTop 10 movies that are hardest to predict:")
print("Movie\tCOD\tBest predictor")
for i in range(1, 11):
    movie = sorted_cods[-i][0]
    cod = sorted_cods[-i][1]
    best_predictor = df.corrwith(df.iloc[:, movie]).sort_values(ascending=False).index[1]
    print(f"{movie}\t{cod:.4f}\t{best_predictor}")
```
